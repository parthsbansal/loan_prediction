## Model Reasoning 

After trying severl models, the two models that performed really well were Linear Regressson and Random Forest Regression. Both models performed equally well. 
However, I decided to use Random Forest Regression to generate loan predictions. Random Forest Regressor is a popular machine learning model that is often used for regression problems, in which the goal is to predict a continuous target variable. It's main pros are its capabilites to capute non-plinear relationships and it also provides an insight to feature importance. If I had more time, I would remove certain less imactful features while doing the predictions. Furthmore, Random Forest Regressor is an ensemble learning method that combines multiple decision trees to reduce overfitting and improve the accuracy of the predictions. This can be particularly useful in high-dimensional datasets where there are many features to consider. 

On the otherhand, it can be expensive. It took way longer to compute predictions than Linear Regression. If our dataset was any larger, Random Forest would not be ideal as it would be computationally expensive!
